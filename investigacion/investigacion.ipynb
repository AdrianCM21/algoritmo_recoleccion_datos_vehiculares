Elijo Hora inicio sumado a minutos

```python
df = pd.read_csv('resul.csv')


print("\nInformación del tipo de datos:")

df['Hora_Inicio'] = pd.to_datetime(df['Hora_Inicio'], format='%H:%M:%S')
df['Hora_Entera'] = df['Hora_Inicio'].dt.hour
df['Minuto_Entrada'] = df['Hora_Inicio'].dt.minute
# df.info()
# Seleccionamos las columnas útiles para el Clustering
# NO usamos 'Hora_Inicio' (texto) ni 'Dia_Semana' (texto).
# Usamos las versiones numéricas.
features = ['Total_Vehiculos', 'Tiempo_Medio_s', 'Ocupacion_Espacial_%', 'Hora_Entera', 'Dia_Semana','Direccion']

X = df[features]
print("\nDatos listos para escalar:")
display(X.head())
```

Escalado de Datos: se preparan los datos y se colocan todas las variables en una escala para su procesado

```python
scaler = StandardScaler() 
X_scaled = scaler.fit_transform(X) 
X_scaled_df = pd.DataFrame(X_scaled, columns=features)

print("Datos escalados correctamente.") display(X_scaled_df.head())
```
